```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop(['deck', 'embark_town', 'class', 'alive', 'who', 'adult_male' ],axis=1, inplace= True)
```

```python
df.info()
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df.dropna(subset=['embarked'], inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df.head()
```

```python
le = LabelEncoder()
```

```python
df['sex'] = le.fit_transform(df['sex'])
```

```python
df['embarked'] = le.fit_transform(df['embarked'])
```

```python
df.head()
```

```python
df['alone'] = df['alone'].astype(int)
```

```python
df.head()
```

```python
x = df.drop('survived', axis= 1)
y= df['survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_test, y_pred)
```

```python
score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

## KNN

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.fit_transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model2 = KNeighborsClassifier(n_neighbors= 5)
```

```python
model2.fit(X_train_scaler, y_train)
```

```python
y_pred = model2.predict(X_test_scaler)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

## Naive Bayse

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model3 = GaussianNB()
```

```python
model3.fit(X_train, y_train)
```

```python
y_pred = model3.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model4 = DecisionTreeClassifier(random_state= 42)
```

```python
y_pred = model4.predict(X_test_scaler)
```
